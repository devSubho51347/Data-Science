```python
import random 
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
li = [random.randint(2,12) for ele in range(100)]
```

```python
li
```

```python
plt.hist(li, bins = 20)
plt.savefig('hist.png')
```

```python
sns.kdeplot(li)
plt.savefig("KernalDensityEstimation.png")
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(nrows = 3,  figsize = (10,20))

ax1.hist(li, bins = 20, color = 'orange')
ax1.set_title("Outcomes of 2 dices thrown simultaneously", size = 15)

ax1.set_xlabel("Dice output")
ax1.set_ylabel("Frequency")

sns.kdeplot(li, ax = ax2, linewidth = 3, color = 'red')

ax2.set_title('KDE plot of Outcomes of 2 dices thrown simultaneously', size = 15)

ax2.set_xlabel("Dice output")
ax2.set_ylabel("Frequency")

sns.kdeplot(li, ax = ax3, linewidth = 3, color = 'blue', cumulative = True)

ax3.set_title('Cumulative Density plot of Outcomes of 2 dices thrown simultaneously', size = 15)

ax3.set_xlabel("Dice output")
ax3.set_ylabel("Cumulative Frequency")

plt.show()

# In order to save subplots
fig.savefig("001911701070_SubhadeepChoudhury.pdf")
```
